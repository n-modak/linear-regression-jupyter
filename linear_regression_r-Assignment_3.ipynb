This notebook demonstrates a simple linear regression analysis using [R] to model Salary based on Years of Experience.

```r
# Load Library
library(ggplot2)
```

```r
# Describe data
filename <- "regression_data.csv"
x <- "YearsExperience"
y <- "Salary"
```

```r
# Read in the data
df <- read.csv(filename)
print(df)
```

```r
# Fit model to data
model <- lm(as.formula(paste(y,"~", x)), data = df)
slope <- coef(model)[2]
intercept <- coef(model)[1]
r <- cor(df[[x]], df[[y]])
pred <- predict(model)
mse <- mean((df[[y]] - pred)^2)
```

```r
# Plot data
ggplot(df, aes_string(x = x, y = y)) +
    geom_point() +
    geom_smooth(method = "lm", se = FALSE, color = "red") + 
    annotate("text", x = 1.5, y = max(df[[y]] +20000),
        label = paste("y =", round(slope, 2), "x +", round(intercept, 2),
                "/nr =", round(r,2), "/nMSE =", round(mse,2)),
        size = 4) +
    labs(title = "Linear Fit",
             x = "x", y = "y") +
    theme_minimal()
ggsave("regression_plot_r.png")
```
